##Activity #1

Derive the 1-dof freedom model for describing the small amplitude motion of the system.

Measuring the displacement of the center of mass of the rotor from its equilibrium position and using the Newtons equations;

\begin{align}
M\ddot{x}+C\dot{x}+Kx=F(t)
\end{align}

Being a linear second order ODE one can solve it explicitly. Since this differential equation was derived by considering small deflections and velocities of the mass it is interesting to find out if the solutions of the differential equation correspond to the physical behaviour of the SMD system for small deflections.

One way of finding the solutions of this differential equation is the Laplace  transform method that we use below.

Let $\omega^2_n=\frac{K}{M}$ and $2 \zeta \omega_n=\frac{C}{M}$. Then the above equations take the form
\begin{align}
\ddot{x}(t) + 2 \zeta \omega_n \dot{x}(t) + \omega^2_n x(t)= \frac{1}{M}F(t).
\end{align}

The parameter $\omega_n$ is called the undamped natural frequency of the system and $\zeta$ is called the damping ratio of the system. The reason for this nomenclature will be apparent when we investigate the solutions of this differential equation.

For a given $f(t)$ that does not grow faster than an exponential function it can be shown, owing to the linearity of the system, that the solution will also not grow
faster than an exponential function. Thus taking Laplace transform
of both sides of and using the linearity property of the Laplace transform we have
\begin{align}
\mathfrak{L}\left\{\ddot{x}(t) + 2 \zeta \omega_n \dot{x}(t) + \omega^2_n x(t)\right\}&=\frac{1}{M}\mathfrak{L}\left\{f(t)\right\},\\
(s^2+2 \zeta \omega_n s + \omega_n^2) X(s) - \dot{x}(0)- ( s + 2\zeta \omega_n)x(0)  &= \frac{1}{M} F(s).
\end{align}
This yields
\begin{align}
X(s)=\frac{1}{s^2+2 \zeta \omega_n s + \omega_n^2} \dot{x}(0)+\frac{ s + 2\zeta \omega_n}{s^2+2 \zeta \omega_n s + \omega_n^2}x(0)
+\frac{1/M}{s^2+2 \zeta \omega_n s + \omega_n^2}F(s).
\end{align}
Since the Laplace is a one-to-one and onto operator its inverse exists and thus the the solution can be uniquely determined to be $x(t)=\mathfrak{L}^{-1}\{X(s)\}$. Using the linearity property of the inverse we thus have

\begin{align}
x(t)&=\dot{x}(0)\mathfrak{L}^{-1}\left\{\frac{1}{s^2+2 \zeta \omega_n s + \omega_n^2} \right\}+x(0)\mathfrak{L}^{-1}\left\{\frac{ s + 2\zeta \omega_n}{s^2+2 \zeta \omega_n s + \omega_n^2}\right\}+\mathfrak{L}^{-1}\left\{\frac{1/M}{s^2+2 \zeta \omega_n s + \omega_n^2}F(s)\right\}.
\end{align}

Observe that the first two terms depend on the initial conditions and do not depend on the forcing while the last term does depend on the forcing but is in depend of the initial conditions. Therefore
we may breakup the solution into two parts such that $x(t)=x_{IC}(t)+x_f(t)$ where

\begin{align}
x_{IC}(t)&=\dot{x}(0)\mathfrak{L}^{-1}\left\{\frac{1}{s^2+2 \zeta \omega_n s + \omega_n^2} \right\}+x(0)\mathfrak{L}^{-1}\left\{\frac{ s + 2\zeta \omega_n}{s^2+2 \zeta \omega_n s + \omega_n^2}\right\}, \\
x_f(t)&=\mathfrak{L}^{-1}\left\{\frac{1/M}{s^2+2 \zeta \omega_n s + \omega_n^2}F(s)\right\}.
\end{align}

We will call the first part two parts given by $x_{IC}(t)$ that depends only on the initial conditions the initial condition response while we will call the last part that depends only on the forcing the forced response.

The inverse Laplace transforms of each of the terms are obtained by expanding the terms in a partial fraction expansion. This depends on the roots of the polynomial
\begin{align}
\Delta(s)&=s^2+2 \zeta \omega_n s + \omega_n^2.
\end{align}

Let us consider the case $0\leq \zeta <1$. In this case the characteristic polynomial is $\Delta(s)=s^2+2 \zeta \omega_n s + \omega_n^2$ of the system.

##Activity #2

Estimate the damped natural frequency, $\omega_d$, and the amplitude of reseonance, $x_{ss_{max}}$, for small amplitude vibrations.

### Damped natural frequency $({x_d})$
Damped natural frequency is given by;
\begin{align}
\omega_d = \omega_n\sqrt{1-\zeta^2}
\end{align}

where;

\begin{align}
\omega_n = \sqrt{\frac{k}{m}}
\end{align}

\begin{align}
\zeta = \frac{c}{2\sqrt{km}}
\end{align}

### Amplitude of Resonance $({x_{ss}}_{max})$

Let us consider a very particular type of forcing that always arises in the presence of roatary parts in a system.
\begin{align}
F(t)=F_0\,\cos{(\omega t)}.
\end{align}
Since the Laplace transform of $F(t)$ is

\begin{align}
F(s)=F_0\left(\frac{s}{s^2+\omega^2}\right)=F_0\left(\frac{s}{(s+i\omega)(s-i\omega)}\right).
\end{align}

Thus it can be shown that the steady state response of the system for sinusoidal forcing is
\begin{align}
x_{ss}(t)=\chi(\omega)\,F_0\,\cos(\omega t +\phi(\omega))\:1(t),
\end{align}
where
\begin{align}
\chi(\omega)=\frac{1/M}{\sqrt{(\omega_n^2-\omega^2)^2+4\zeta^2\omega^2\omega_n^2}}
\end{align}

and

\begin{align}
\phi(\omega)=\arctan{\left(\frac{2\zeta\omega\omega_n}{\omega_n^2-\omega^2}\right)}.
\end{align}

Thus we see that the steady state response to sinusoidal forcing is also sinusoidal with frequency of oscillation equal to that of the forcing frequency $\omega$.

However it differs from the forcing in two distinct ways. One is that the amplitude of the steady state solution is $\chi(\omega)$ times the magnitude of the forcing while the other is that the phase of the steady state solution is shifted by $\phi(\omega)$ from the phase of the forcing.

It is also important to note that the amplitude magnification and the phase shift depend on the forcing frequency $\omega$.

We will see that this observation has crucial implications for vibration analysis and control system design.

One can show that $\chi(\omega)$ is maximum when
\begin{align}
\omega=\omega_r=\omega_n\sqrt{1-2\zeta^2},
\end{align}
and that the maximum amplitude of vibration is given by
\begin{align}
{x_{ss}}_{max}=\frac{F_0/M}{2\zeta\omega_n^2\sqrt{1-\zeta^2}}.
\end{align}


${F_0}$is the amplitude of the external force,

${c}$ is the damping coefficient,

${ω_n}$ is the natural frequency.

## 1. Unforced Response

The equation of motion for an unforced damped harmonic oscillator is:

\begin{align}
m\ddot{x}(t) + c \dot{x}(t) + k x(t) = 0
\end{align}

Key Definitions:
\begin{align*}
m & : \text{Mass (kg)} \\
c & : \text{Damping coefficient (Ns/m)} \\
k & : \text{Spring constant (N/m)} \\
\omega_n & = \sqrt{\frac{k}{m}} : \text{Natural frequency (rad/s)} \\
\zeta & = \frac{c}{2\sqrt{mk}} : \text{Damping ratio}
\end{align*}

Solution:

For the underdamped case $( \zeta < 1 )$,

the solution is:

\begin{align}
x(t) = e^{-\zeta \omega_n t} \left[ A \cos(\omega_d t) + B \sin(\omega_d t) \right]
\end{align}

where  $\omega_d = \omega_n \sqrt{1 - \zeta^2} \$;


Using initial conditions:

\begin{align}
x(0) = x_0, \quad \dot{x}(0) = v_0
\end{align}


we find:
\begin{align}
A = x_0, \quad B = \frac{v_0 + \zeta \omega_n x_0}{\omega_d}.
\end{align}


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# System parameters
mass = 1.0  # kg
k = 100.0  # N/m (spring constant)
c = 2.0  # Ns/m (damping coefficient)

# Initial conditions
x0 = 0.1  # Initial displacement (meters)
v0 = 0.0  # Initial velocity (m/s)

# Natural frequency and damping ratio
omega_n = np.sqrt(k / mass)
zeta = c / (2 * np.sqrt(mass * k))

# Define the system of ODEs
def unforced_ode(t, y):
    x, v = y
    dxdt = v
    dvdt = -2 * zeta * omega_n * v - omega_n**2 * x
    return [dxdt, dvdt]

# Time span
t_span = (0, 5)  # 5 seconds
t_eval = np.linspace(t_span[0], t_span[1], 500)

# Solve the ODE
sol = solve_ivp(unforced_ode, t_span, [x0, v0], t_eval=t_eval)

# Plot the unforced response
plt.figure(figsize=(10, 6))
plt.plot(sol.t, sol.y[0], label="Displacement (x)")
plt.title("Unforced Response")
plt.xlabel("Time (s)")
plt.ylabel("Displacement (m)")
plt.grid()
plt.legend()
plt.show()
```

## 2. Forced Response for Three Rotational Speeds

The equation of motion for a forced damped harmonic oscillator is:

\begin{align}
m \ddot{x}(t) + c \dot{x}(t) + k x(t) = F_0 \cos(\omega t)
\end{align}

Steady-State Solution:

\begin{align}
x(t) = X(\omega) \cos(\omega t - \phi)
\end{align}

where:
\begin{align}
X(\omega) = \frac{F_0}{\sqrt{\left( k - m \omega^2 \right)^2 + \left( c \omega \right)^2}}
\end{align}

and:
\begin{align}
\phi = \tan^{-1} \left( \frac{c \omega}{k - m \omega^2} \right).
\end{align}

The total response is a combination of transient and steady-state responses. The transient response diminishes over time due to damping.

```python
# Forcing function parameters
motor_speeds = [5, 10, 15]  # rad/s (example rotational speeds)
F0 = 10.0  # N (forcing amplitude)

# Define the forced ODE
def forced_ode(t, y, omega):
    x, v = y
    dxdt = v
    dvdt = (F0 * np.cos(omega * t) - 2 * zeta * omega_n * v - omega_n**2 * x) / mass
    return [dxdt, dvdt]

# Plot forced responses for different motor speeds
plt.figure(figsize=(10, 6))
for omega in motor_speeds:
    sol = solve_ivp(forced_ode, t_span, [0, 0], t_eval=t_eval, args=(omega,))
    plt.plot(sol.t, sol.y[0], label=f"ω = {omega} rad/s")

plt.title("Forced Response for Different Motor Speeds")
plt.xlabel("Time (s)")
plt.ylabel("Displacement (m)")
plt.grid()
plt.legend()
plt.show()
```

## 3. Theoretical Frequency Response

The transfer function of the system is:
\begin{align}
H(\omega) = \frac{1}{k - m\omega^2 + j c\omega}
\end{align}

The magnitude of the frequency response is:

\begin{align}
|H(\omega)| = \frac{1}{\sqrt{\left( k - m \omega^2 \right)^2 + \left( c \omega \right)^2}}
\end{align}


The system's response depends on the input frequency $\omega$. This function characterizes how the system reacts to different forcing frequencies.

```python
# Frequency response
frequencies = np.linspace(0.1, 20, 500)  # rad/s
omega_d = frequencies
H = F0 / np.sqrt((k - mass * omega_d**2)**2 + (c * omega_d)**2)

# Plot the frequency response
plt.figure(figsize=(10, 6))
plt.plot(frequencies, H)
plt.title("Frequency Response of the System")
plt.xlabel("Frequency (rad/s)")
plt.ylabel("Amplitude (m)")
plt.grid()
plt.show()
```

##Activity #5

Design a tuned mass vibration absorber for the system and justify your design using simulations.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# System parameters
M = 10  # Primary mass (kg)
K = 1000  # Primary stiffness (N/m)
C = 10  # Primary damping (Ns/m)
omega_f = 5  # Forcing frequency (rad/s)

# Absorber parameters
m = 1  # Absorber mass (kg)
k = m * omega_f**2  # Absorber stiffness (N/m)
c = 0.05 * 2 * np.sqrt(k * m)  # Absorber damping (Ns/m)

# Equations of motion
def equations(t, y):
    x_M, v_M, x_m, v_m = y
    F = np.sin(omega_f * t)  # Harmonic force
    dx_M = v_M
    dv_M = (F - C * v_M - K * x_M - k * (x_M - x_m) - c * (v_M - v_m)) / M
    dx_m = v_m
    dv_m = (-k * (x_m - x_M) - c * (v_m - v_M)) / m
    return [dx_M, dv_M, dx_m, dv_m]

# Time range for simulation
t_span = (0, 10)
t_eval = np.linspace(0, 10, 1000)
y0 = [0, 0, 0, 0]  # Initial conditions

# Solve the system with the absorber
solution = solve_ivp(equations, t_span, y0, t_eval=t_eval)
x_M = solution.y[0]  # Displacement of primary mass
x_m = solution.y[2]  # Displacement of absorber mass

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(t_eval, x_M, label='Primary Mass (with absorber)', color='blue')
plt.plot(t_eval, x_m, label='Absorber Mass', color='red', linestyle='--')
plt.title("Vibration Response with Tuned Mass Absorber")
plt.xlabel("Time (s)")
plt.ylabel("Displacement (m)")
plt.legend()
plt.grid(True)
plt.show()
```
